```python
import torch
import matplotlib.pyplot as plt
import torchvision.utils as vutils

# Load
run_name = "bumbling-thunder-13"  # replace with your actual run name
data = torch.load(f"logged_files/distillation/waterbirds/dinov2_vitb/{run_name}/data.pth", weights_only=False)
print(data.keys())  # voir ce qu'il y a dedans

images = data["images"]  # (N, C, H, W)
print(f"Shape: {images.shape}")

# Normalize to [0, 1] for display
def normalize(img):
    img = img - img.min()
    img = img / img.max()
    return img

# Display grid
grid = vutils.make_grid(normalize(images), nrow=4, padding=2)
plt.figure(figsize=(12, 12))
plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
plt.axis("off")
plt.title("Distilled Images")
plt.show()
print("Saved to distilled.png")
```
